```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
from bs4 import BeautifulSoup
import time

def wait_for_loading_to_finish(driver, timeout=20):
    WebDriverWait(driver, timeout).until(
        lambda d: d.find_element(By.ID, "ctl00_UpdateProgressMaster").get_attribute("aria-hidden") == "true"
    )

def scrape_hctx(start_date, end_date, doc_type):
    url = "https://www.cclerk.hctx.net/Applications/WebSearch/RP_R.aspx?ID=PtRyJzbPPV9CWT5QJ8WvKEFVAr+pwQL/1XGVmC/aHdfP+DIXvYNpdnX9R8yeQC2XXgkUgoYdO2y3PRZ2zTfJGhb/6xRICV2VPFSofHebrWv7hxYW2UwdAi1h77pFQpJntC2qr8qMihkZHUM/4mzhH1C9e7qmkcSdI3sSSVXPAtA="
    driver = webdriver.Chrome()
    driver.get(url)
    
    try:
        # Wait for the date input fields to be present
        wait = WebDriverWait(driver, 10)
        
        # Fill in the start date
        start_date_input = wait.until(EC.presence_of_element_located((By.ID, "ctl00_ContentPlaceHolder1_txtDateN")))
        start_date_input.send_keys(start_date)
        
        # Fill in the end date
        end_date_input = driver.find_element(By.ID, "ctl00_ContentPlaceHolder1_txtDateTo")
        end_date_input.send_keys(end_date)
        
        # Fill in the document type
        doc_type_input = driver.find_element(By.ID, "ctl00_ContentPlaceHolder1_txtType")
        doc_type_input.send_keys(doc_type)
        
        # Click the search button
        search_button = driver.find_element(By.ID, "ctl00_ContentPlaceHolder1_btnSearch")
        search_button.click()
        time.sleep(2)
        
        all_data = []  # To store all rows from all pages
        
        while True:
            print("Waiting for results to load...")
            wait_for_loading_to_finish(driver)
            
            # Wait for the table to appear
            try:
                response = driver.page_source
                soup = BeautifulSoup(response, 'html.parser')
                table = soup.find('table', {'id': 'itemPlaceholderContainer'})
                df = pd.read_html(str(table))[0]  # Extract table using BeautifulSoup
                all_data += list(df["Legal Description"].dropna())
            except Exception:
                print("No records found. Stopping.")
                break
            
            # Try clicking the next button using JavaScript
            try:
                next_button = driver.find_element(By.ID, "ctl00_ContentPlaceHolder1_BtnNext")
                if next_button.is_enabled():
                    print("Clicking Next button...")
                    next_button.click()
                    time.sleep(2)
                    wait_for_loading_to_finish(driver)
                else:
                    print("Next button is disabled. Stopping.")
                    break
            except Exception as e:
                print("Error clicking Next button:", e)
                break
        
    finally:
        driver.quit()
    
    return all_data


# Example usage
start_date = "03/17/2024"
end_date = "03/21/2024"
doc_type = "NOTICE"
df = scrape_hctx(start_date, end_date, doc_type)
print(df)
```

```python
import pandas as pd
from lxml import html

def extract_raw_text_from_tr(tr_element):
    """Extract raw text from a <tr> element, handling nested tables."""
    row_data = []
    for td in tr_element.xpath('.//td'):  # Extract all td elements inside tr
        # Check if there is a nested table
        nested_table = td.xpath('.//table')
        if nested_table:
            # If nested table found, recursively extract its raw text
            nested_text = extract_raw_text_from_table(nested_table[0])
            # Flatten the nested text into a single list if necessary
            row_data.append(" | ".join(str(item) for sublist in nested_text for item in sublist))
        else:
            # Otherwise, just extract the text from the <td> element
            row_data.append(td.text_content().strip())
    return row_data

def extract_raw_text_from_table(table_element):
    """Extract raw text from all rows in a table."""
    raw_data = []
    for tr in table_element.xpath('.//tr'):  # Get all <tr> tags within the table
        row = extract_raw_text_from_tr(tr)
        if row:
            raw_data.append(row)
    return raw_data

def convert_table_to_dataframe(html_content):
    """Convert an HTML table to a pandas DataFrame, extracting raw text."""
    tree = html.fromstring(html_content)
    table = tree.xpath('//table')[3]  # Assuming we want the first table in the HTML
    raw_data = extract_raw_text_from_table(table)
    
    # Ensure all rows have the same length as the longest row
    max_columns = max(len(row) for row in raw_data)
    for row in raw_data:
        while len(row) < max_columns:  # Add empty strings if needed to match max_columns
            row.append('')
    
    # Check if the table has data and use the first row as column headers
    if len(raw_data) > 1:
        df = pd.DataFrame(raw_data[1:], columns=raw_data[0])  # First row as header
    else:
        df = pd.DataFrame(raw_data)  # If no header, just make it a regular DataFrame
    return df

# Example usage
html_content = response  # Replace with your actual HTML content

# Convert HTML table to DataFrame
df = convert_table_to_dataframe(html_content)

# Show DataFrame
print(df)
```
